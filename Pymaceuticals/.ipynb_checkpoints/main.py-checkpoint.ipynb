## Pymaceuticals, Inc.

Analysis
-Of the ten drugs evaluated, Capomulin and Ramicane demonstrated tumor growth inhibition in mouse models while the rest did not show significant tumor inhibition compared with the placebo group.
-Capomulin and Ramicane demonstrated reduced metastatic spread during the treatment in comparison to placebo and other drugs as well.
-Capomulin and Ramicane demonstrated significantly improved survival rates in comparison with the placebo and other drugs.
-Capomulin and Ramicane demonstrated significantly tumor shrinkage. 

```python
#Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import sem
```

```python
#Read the datafiles to pandas dataframes
datafile1="raw_data/mouse_drug_data.csv"
datafile2="raw_data/clinicaltrial_data.csv"
df1=pd.read_csv(datafile1)
df2=pd.read_csv(datafile2)
print(df1.head())
print(df2.head())
```

```python
#Merge the two dataframes
df=pd.merge(df1, df2, on="Mouse ID")
df.head()
```

## Tumor Response to Treatment

```python
#Generate a pivot table to summarize the data for tumor volume and metastatic sites over time
df_pivot=df.pivot_table(index=['Drug', 'Timepoint'],values=['Tumor Volume (mm3)','Metastatic Sites'],aggfunc=np.mean)
df_pivot.head()
```

```python
#Generate a pivot table to generate the standard error data for tumor volume and metastatic sites over time
std_error=df.pivot_table(index=['Drug', 'Timepoint'],values=['Tumor Volume (mm3)','Metastatic Sites'],aggfunc=sem)
std_error.head()
```

```python
#Get unique drug names
drug_index=df_pivot.index.get_level_values(0).unique()
drug_index
```

```python
#Get % tumor volume change from baseline time point
#Get the list of tumor volume change over time for each drug
tumor_vol_change_list=[]

#loop through all the drugs
for drug in drug_index:
    
    #Tumor volume at timepoint 0
    tumor_vol_t0=df_pivot.loc[drug].loc[0]['Tumor Volume (mm3)']
    
    #Number of timepoints
    num_timepoints=len(df_pivot.loc[drug]['Tumor Volume (mm3)'].index)
    
    #Loop through each timepoint
    for i in range(num_timepoints):
        
        #Each timepoint
        tp=5*i
        #Tumor volume at each timepoint
        tumor_vol_tp=df_pivot.loc[drug].loc[tp]['Tumor Volume (mm3)']
        #Tumor volume change relative to timepoint 0
        tumor_vol_change=(tumor_vol_tp-tumor_vol_t0)/tumor_vol_t0*100
        #Generate a list of tumor volume change
        tumor_vol_change_list.append(tumor_vol_change)

#Convert the tumor change list to dataframe
tumor_vol_change_df=pd.DataFrame({"Tumor Volume Change":tumor_vol_change_list}, index=df_pivot.index)
tumor_vol_change_df.head()
```

```python
#Make a dataframe to show the tumor volume over treatment time
#Tumor volume data for the first drug (d1)
tumor_vol_d1=df_pivot.loc['Capomulin']['Tumor Volume (mm3)']

#Loop through the remaining drugs
for drug in drug_index[1:]:
    #Tumor volume data for drug x (dx)
    tumor_vol_dx=df_pivot.loc[drug]['Tumor Volume (mm3)']
    #Combine the dataframes
    tumor_vol_dall=pd.concat([tumor_vol_d1,tumor_vol_dx], axis=1) 
    #Assign the combined dataframe to  tumor_vol_d1
    tumor_vol_d1=tumor_vol_dall

#Assign the column names
tumor_vol_dall.columns=drug_index

tumor_vol_dall.head()
```

```python
#Plot a line plot of tumor volume against the treatment time (days)
x_axis=tumor_vol_dall.index
color_panel=['r','g','b','y','r','g','b','y','r','g']
filled_markers = ['o', 'v', '^', '<', '>', '8', 's', 'p', '*', 'h', 'H', 'D', 'd', 'P', 'X']
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(8,6))
for i,drug in enumerate (drug_index):
    plt.errorbar(x_axis, tumor_vol_dall[drug], yerr=std_error.loc[drug]['Tumor Volume (mm3)'], color=color_panel[i], marker=filled_markers[i], markersize=5, dashes=[5,5,5,5], linewidth=0.8)
    #plt.errorbar(x_axis, tumor_vol_dall[drug], yerr=std_error.loc[drug]['Tumor Volume (mm3)'], color=color_panel[i], marker=filled_markers[i], markersize=5, linestyle='--', linewidth=0.8)

# Set x and y limits
plt.xlim(0, max(x_axis))
plt.ylim(20, 80)

# Set a Title and labels
plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(drug_index, loc='best')
plt.grid(True)
# Save our graph and show the grap
#plt.savefig("tumor_vol_change.png")
plt.tight_layout()
plt.show()

#standard_errors = [sem(s) for s in samples]
# Generate an errorbar plot
# The 'yerr' argument refers to the error bar for each sample, oriented to the y-axis
```

## Metastatic Response to Treatment

```python
#Plot a line plot of metastastic sites against the treatment time (days)
x_axis=tumor_vol_dall.index
color_panel=['r','g','b','y','r','g','b','y','r','g']
filled_markers = ['o', 'v', '^', '<', '>', '8', 's', 'p', '*', 'h', 'H', 'D', 'd', 'P', 'X']
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(8,6))
for i,drug in enumerate (drug_index):
    plt.errorbar(x_axis, df_pivot.loc[drug,'Metastatic Sites'], yerr=std_error.loc[drug]['Metastatic Sites'], color=color_panel[i], marker=filled_markers[i], markersize=5, linestyle='--', linewidth=0.8)

# Set x and y limits
plt.xlim(0, max(x_axis))
#plt.ylim(20, 80)

# Set a Title and labels
plt.title("Metastatic Spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")
plt.legend(drug_index, loc='best')
# Save our graph and show the grap
#plt.savefig("tumor_vol_change.png")
plt.grid(True)
plt.tight_layout()
plt.show()

#standard_errors = [sem(s) for s in samples]
# Generate an errorbar plot
# The 'yerr' argument refers to the error bar for each sample, oriented to the y-axis
```

## Survival Rates

```python
survival_count=df.groupby(["Drug","Timepoint"])['Tumor Volume (mm3)'].count()
survival_count.head()
```

```python
survival_d1=survival_count['Capomulin']
for drug in drug_index[1:]:
    survival_dx=survival_count[drug]
    survival_summary=pd.concat([survival_d1,survival_dx], axis=1) 
    survival_d1=survival_summary
survival_summary.columns=drug_index
survival_summary.head()
```

```python
# Tells matplotlib that we want to make a scatter plot
# The size of each point on our plot is determined by their x value
x_axis=survival_summary.index
color_panel=['r','g','b','y','r','g','b','y','r','g']
filled_markers = ['o', 'v', '^', '<', '>', '8', 's', 'p', '*', 'h', 'H', 'D', 'd', 'P', 'X']
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(8,6))
for i,drug in enumerate (drug_index):
    num_t0=survival_summary[drug][0]
    plt.plot(x_axis,survival_summary[drug]/num_t0*100, marker=filled_markers[i], markersize=5, c=color_panel[i], linestyle='--', linewidth=0.8)
    #plt.scatter(x,survival_summary[drug]/num_t0, marker=filled_markers[i], facecolors=color_panel[i], edgecolors="black",markersize=5, linestyle='--', linewidth=0.8)

# Use df.plot() to set up our bar graph
#new_df.plot(x=new_df["State"], kind="bar", figsize=(20, 3), title="Average Rain per State")

# Set x and y limits
plt.xlim(0, max(x_axis))
#plt.ylim(20, 80)

# Set a Title and labels
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.legend(drug_index, loc='best')
# Save our graph and show the grap
#plt.savefig("tumor_vol_change.png")
plt.grid(True)
plt.tight_layout()
plt.show()
```
